```python
hr_emotion_labels = pd.merge(emotion_label_P1[['timestamp','arousal','valence']], sensor_all_P1['E4_HR.csv'][['timestamp','value']], on='timestamp')

fig, ax1 = plt.subplots()

ax1.plot(hr_emotion_labels['timestamp'],hr_emotion_labels['value'],label="heart rate",color='red')
ax1.legend()
ax1.set_title("Arousal and Heart Rate")

ax2 = ax1.twinx()
ax2.plot(hr_emotion_labels['timestamp'],hr_emotion_labels['arousal'],label="arousal")
ax2.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(hr_emotion_labels['timestamp'],hr_emotion_labels['value'],label="heart rate",color='red')
ax1.legend()
ax1.set_title("Valence and Heart Rate")

ax2 = ax1.twinx()
ax2.plot(hr_emotion_labels['timestamp'],hr_emotion_labels['valence'],label="valence")
ax2.legend()
plt.show()
```

Valence = positive (high) to negative (low). Here, positive valence signifies a decreased heart rate (which can signify calmness (maybe after winning an arguement, or the end of an arguement). However, in the start, valence and heart rate are correlated (which can signify that a positive valence gives rise to the heart rate (such as positive emotions). It infers that, the heart rate can rise or fall with a positive valence, depending on the context. In this case, the heart rate tends to follow an inverse relationship with positive valence (maybe due to the participant being comfortable with the debate (and not excited anymore)). The start of the debate can have a "settling down" period.

Arousal = alert (high) vs calm (low). Here in the start of the conversation, arousal and heart rate are not positively correlated (i.e, a self reported low arousal or high arousal result in the opposite affect on the heart rate) and there also may be a time-lag. But the participant is still able to rate *changes* in arousal and heart rate well for the start of the conversation. Afterwards, arousal tends to stay the same but there are changes in heart rate. Valence was better at describing these changes. Then near the end of the conversation, the participant is able to correctly capture the changes in arousal and heart rate (but with a time lag). Important insights: start of the conversation tends to have an inverse affect on what the participant percieves, but this settles down and tends to become in sync later. The middle of the converstaion tends to have variations, but no percievable change in arousal. And the end of the conversation tends to be in sync with the heart rate. This can be described as the participant was watching himself after feeling those emotions (which is the general way of collecting emotion labels also in DyMand). As the total conversation was more than 14 minutes long, the participant may or may not have been tired of rating his emotions (or noticing his emotions) and was more concentrated at the beginning and end of the conversation.
